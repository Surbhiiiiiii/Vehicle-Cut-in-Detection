```python
# pip install opencv-python
# Resizing all the input training images.
import os
import cv2

def resize_image(image_path, output_path, size=(512, 512)):
    """Resize the image to the given size and save it to the output path."""
    img = cv2.imread(image_path)
    if img is not None:
        resized_img = cv2.resize(img, size, interpolation=cv2.INTER_AREA)
        cv2.imwrite(output_path, resized_img)

def resize_images_in_dataset(dataset_dir, output_dir, size=(512, 512)):
    """Resize all images in the dataset directory and save them to the output directory."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for root, _, files in os.walk(dataset_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
                image_path = os.path.join(root, file)
                relative_path = os.path.relpath(image_path, dataset_dir)
                output_path = os.path.join(output_dir, relative_path)

                # Ensure the output directory exists
                os.makedirs(os.path.dirname(output_path), exist_ok=True)
                resize_image(image_path, output_path, size)

if __name__ == "__main__":
    dataset_directory =   "C:/Users/hp/Downloads/New folder (2)/idd_temporal_train_1"# replace with the path to your dataset
    output_directory = "C:/Users/hp/Downloads/New folder (2)/idd_temporal_train_1"  # replace with the path to save resized images
    resize_images_in_dataset(dataset_directory, output_directory)
```

```python
import os
from PIL import Image
import imagehash

def find_duplicate_images(directory):
    image_hashes = {}
    duplicates = []

    # Traverse the directory and process images
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                filepath = os.path.join(root, filename)
                try:
                    # Open the image file
                    with Image.open(filepath) as img:
                        # Compute the hash of the image
                        hash_value = imagehash.average_hash(img)

                    # Check if the hash already exists
                    if hash_value in image_hashes:
                        duplicates.append(filepath)
                    else:
                        image_hashes[hash_value] = filepath

                except Exception as e:
                    print(f"Error processing {filepath}: {e}")

    return duplicates

def remove_duplicates(duplicate_files):
    for file in duplicate_files:
        try:
            os.remove(file)
            print(f"Removed {file}")
        except Exception as e:
            print(f"Error removing {file}: {e}")

if __name__ == "__main__":
    dataset_directory = 'C:/Users/hp/Downloads/idd_temporal_val_1'
    duplicate_files = find_duplicate_images(dataset_directory)
    print(f"Found {len(duplicate_files)} duplicate images.")
    remove_duplicates(duplicate_files)
```
